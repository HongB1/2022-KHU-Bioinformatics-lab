```python
# 1. ax.bar Basics
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

n_data = 10
data = np.random.uniform(10, 100, (n_data,)) # random.uniform(low=0.0, high=1.0, size=None)
data_idx = np.arange(n_data) # numpy.arange([start, ]stop, [step, ]dtype=None, *, like=None)

fig, ax = plt.subplots(figsize=(10, 7))
ax.tick_params(labelsize=15)

ax.bar(data_idx, data)
```

```python
n_data = 10
# data = np.random.uniform(10, 100, (n_data,))
background = 100*np.ones(n_data)
data_idx = np.arange(n_data)

fig, ax = plt.subplots(figsize=(10, 7))
ax.tick_params(labelsize = 15)
ax.bar(data_idx, background, facecolor='whitesmoke', hatch='/', edgecolor='silver')
ax.bar(data_idx, data, facecolor='cornflowerblue')
data_labels = ['class ' + str(i) for i in range(n_data)]
ax.bar(data_idx, data) # tick_labels 사용하지 않음
ax.set_xticks(data_idx)
ax.set_xticklabels(data_labels, rotation=-30, ha='left')

major_yticks = np.arange(0, 101, 20)
major_yticklabels = [str(p) + '%' for p in major_yticks]
minor_yticks = np.arange(0, 101, 5)

ax.set_yticks(major_yticks)
ax.set_yticklabels(major_yticklabels)
ax.set_yticks(minor_yticks, minor=True)

ax.grid(axis='y', which='major')
ax.grid(axis='y', which='minor', linestyle=':')
```

```python
# Bar Plot with Sorting # IMPORTANT 이거 꼭 쓰기 HLA 순서대로
np.random.seed(0)

n_data = 10
data = np.random.uniform(0, 10, n_data)
data_idx = np.arange(n_data)
data_labels = np.array(['class ' + str(i+1) for i in range(n_data)])

sort_idx = np.argsort(data) # np.sort(data)[::-1]
data_sort = data[sort_idx]
data_labels_sort = data_labels[sort_idx]

fig, axes = plt.subplots(2, 1, figsize=(10, 12))
axes[0].bar(data_idx, data, tick_label=data_labels)
axes[1].bar(data_idx, data_sort, tick_label=data_labels_sort)
```

```python
gdp_idx = np.arange(len(gdp))
gdp_idx
idx_sort = np.argsort(gdp)
idx_sort # 최소값의 인덱스부터 -> 최댓값의 인덱스까지...
gdp_sort = gdp[idx_sort] # 인덱스를 사용하여 최솟값부터 최댓값까지...
countries_sort = countries[idx_sort]
countries
gdp_p_sort = gdp_p[idx_sort]
```
