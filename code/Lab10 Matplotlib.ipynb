````python
# Exercise 4-01
### data setting
xticklabels = np.array(['WHITE', 'BLACK', "GRAY", 'RED', 'GREEN', 'BLUE','YELLOW','BROWN','OTHER'])
data = np.array([47.19, 15.82, 14.32, 7.30, 2.64, 5.64, 0.84, 5.38, 0.73])
data_idx = np.array(len(data))
background = 55*np.ones(data_idx)
xticks = np.arange(len(data))

fig, ax = plt.subplots(figsize = (20,10))

### background plotting
ax.bar(xticks, background, facecolor="k", alpha=0.05, hatch='//////', edgecolor='k')

### bar plotting
rects = ax.bar(xticks, data, facecolor='cornflowerblue')

### anotation
for rect_idx, rect in enumerate(rects):
    height = rect.get_height()
    width = rect.get_width()
    x = rect.get_x()

    ax.text(x + width/2, 55, f'{height}%', rotation=-50, fontsize=30, ha='right', va='bottom')

### tick & ticklabel customizing
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels, rotation=40, fontsize=30)
ax.tick_params(axis='y', labelsize=0, size=0)
ax.tick_params(axis='x', labelsize=0, size=0)

### spine customizing
for spine_idx, spine_loc in enumerate(ax.spines):
    ax.spines[spine_loc].set_visible(False)
````
