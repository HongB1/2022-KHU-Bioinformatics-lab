### 2-4. `ax.twinx`(Different Y Values)

````python
PI = np.pi
t = np.linspace(0.01, 5*PI, 100)
sin = np.sin(t)
exp = np.exp(t)

fig = plt.figure(figsize=(7, 5))
ax1 = fig.add_subplot()
ax1.plot(t, sin)
ax1.plot(t, exp)
````

````python
PI = np.pi
t = np.linspace(0.01, 5*PI, 100)
sin = np.sin(t)
exp = np.exp(t)

fig = plt.figure(figsize=(7, 5))
ax1 = fig.add_subplot()
ax1.plot(t, sin)
ax2 = ax1.twinx()
ax2.plot(t, exp)
````

````python
fig = plt.figure(figsize=(5, 5))
ax1 = fig.add_subplot()
ax2 = ax1.twinx()
ax1.set_xlim([0, 100])
ax1.set_ylim([0, 100])
ax2.set_ylim([0, 0.1])
````

````python
fig = plt.figure(figsize=(5, 5))
ax1 = fig.add_subplot()
ax2 = ax1.twinx()
ax1.set_xlim([0, 100])
ax1.set_ylim([0, 100])
ax1.set_title('Twinx Graph', fontsize=20)
ax1.set_ylabel('Data1', fontsize=15)
ax2.set_ylim([0, 0.1])
ax2.set_ylabel('Data2', fontsize=15)
````

````python
# fig.suptitle and ax.set_title with Many Axes
ax_title_list = [f'title of {str(i)}' for i in range(4)]
fig, axes = plt.subplots(2, 2, figsize=(7, 7))
fig.suptitle('Title of a Figure', fontsize=20, fontfamily='monospace')

for ax_idx, ax in enumerate(axes.flat):
    ax.set_title(ax_title_list[ax_idx],
                 fontsize=20,
                 fontfamily='serif')
# fig.tight_layout() 은 suptitle을 무시하기 때문에 이렇게 겹쳐지는 경우에는 쓰면 안됨
fig.subplots_adjust(bottom=0.05, top=0.85, hspace=0.3)
````
