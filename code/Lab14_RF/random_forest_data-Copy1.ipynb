```python
# chrom + chromStart + chromEnd -> Pos (chr:####-####)
# Histone.bed -> Histone
# 동일한 gene에서 같은 종류의 histone이 발견되면 distance가 적은 걸로, distance가 같으면 HsignalValue가 큰 값을 남기기

def reshape_df(df):
    df['Pos'] = 'tmp'
    
    for idx in range(len(df)):
        df['Pos'][idx] = f'{df["chr"][idx]}:{df["start"][idx]}-{df["end"][idx]}'
        df['Histone'][idx] = df['Histone'][idx].split('.')[0].split('_')[0]
    
    result = df.reindex(columns = ['Pos', 'gene_name', 'distance with gene',
                                  'Histone', 'HsignalValue'])
    return result
```

```python
EZH2 = reshape_df(ezh2)
EP300 = reshape_df(ep300)
```

```python
from tqdm.notebook import tqdm
```

```python
## 동일한 유전자에서 같은 histone이 발견되면 distance 적은 걸로, 같으면 HsignalValue가 큰 값을 남기기. 
## 같으면 둘중 하나 없애
## list에 넣어서 drop(list)

rm_list = []
for gene in tqdm(concatDF['gene_name'].unique()):
    tmp_df = concatDF[concatDF['gene_name']==gene]
    for idx in list(tmp_df.index):
        for i in list(tmp_df.index):
            if (idx < i) & (tmp_df.loc[idx, "Histone"]==tmp_df.loc[i, "Histone"]) & (tmp_df.loc[idx,'enhancer']==tmp_df.loc[i,'enhancer']):
                if tmp_df.loc[idx, 'distance with gene'] < tmp_df.loc[i,'distance with gene']:
                    rm_list.append(i)
                elif tmp_df.loc[idx, 'distance with gene'] > tmp_df.loc[i,'distance with gene']:
                    rm_list.append(idx)
                elif tmp_df.loc[idx, 'distance with gene'] == tmp_df.loc[i,'distance with gene']:
                    if tmp_df.loc[idx, 'HsignalValue'] < tmp_df.loc[i, 'HsignalValue']:
                        rm_list.append(idx)
                    elif tmp_df.loc[idx, 'HsignalValue'] >= tmp_df.loc[i, 'HsignalValue']:
                        rm_list.append(i)
                        
rm_list = list(set(rm_list))
len(rm_list)
```

```python
concatDF = concatDF.drop(rm_list)
```

```python
concatDF = concatDF.reset_index(drop=True)
concatDF
```
