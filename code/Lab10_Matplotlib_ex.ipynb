```python
def bop_plot(dataset, t_year, ax):
    t_year_data = get_year_data(dataset, t_year).astype(np.int)
    month_dict = {m:None for m in range(1, 1 + 12)} 
    for data_idx, data in enumerate(t_year_data):
        Y, M, D, price = data 
        if month_dict[M] is None:
            month_dict[M] = [data_idx, 
                            '-'.join(str(i)for i in [Y, M, D])]
    
    first_day_label = np.array(list(month_dict.values()))
    x_arange = np.arange(t_year_data.shape[0])
    ax.plot(x_arange, t_year_data[:,-1]) 
    ax.set_xticks(first_day_label[:,0].astype(np.int))
    ax.set_xticklabels(first_day_label[:,1], rotation=-30, ha='left')
    ax.set_title("BOP data({})".format(t_year), fontsize=30)
    ax.tick_params(labelsize=20) 
    ax.grid()
```

```python
# 1. ax.barh
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
data = np.array([1, 2, 3, 4, 5])
data_idx = np.arange(len(data))

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 7))
axes[0].tick_params(labelsize=15)
axes[1].tick_params(labelsize=15)

axes[0].bar(data_idx, data)
axes[1].barh(data_idx, data)

axes[1].invert_yaxis()
```

```python
# 2. ax.barh Examples
np.random.seed(0)

n_data = 10
data = np.random.uniform(10, 100, (n_data,))
data_idx = np.arange(n_data)
data_label = ['class ' + str(i+1) for i in range(n_data)]
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 7))

major_yticks = np.arange(0, 101, 20)
minor_yticks = np.arange(0, 101, 5)

axes[0].bar(data_idx, data)
axes[0].set_xticks(data_idx)
axes[0].set_xticklabels(data_label, rotation=30, ha='left')
axes[0].set_yticks(major_yticks)
axes[0].set_yticklabels(f'{str(i)}%' for i in major_yticks)
axes[0].set_yticks(minor_yticks, minor=True)
axes[0].grid(axis='y', which='major')
axes[0].grid(axis='y', which='minor', linestyle=':')


axes[1].barh(data_idx, data)
axes[1].invert_yaxis()
axes[1].set_yticks(data_idx)
axes[1].set_yticklabels(data_label, rotation=30, va='top')
axes[1].set_xticks(major_yticks)
axes[1].set_xticklabels(f'{str(i)}%' for i in major_yticks)
axes[1].set_xticks(minor_yticks, minor=True)
axes[1].grid(axis='x', which='major')
axes[1].grid(axis='x', which='minor', linestyle=':')


```

```python
# ax.hist Returns rects로 응용하기
n_data = 500
data = np.random.normal(0, 1, (n_data,))

fig, ax = plt.subplots(figsize=(10, 8))
ax.tick_params(labelsize=15, length=10, width=3)
freqs, bin_edges, rects = ax.hist(data, facecolor='skyblue', linewidth=3)
print(freqs) # bar들의 height -> 얼마나 많이 들어 있는 지
print(bin_edges) # 각 부분
print(rects) # bar object 10개가 각각 들어있음

for rect in rects:
    x = rect.get_x()
    width = rect.get_width()
    height = rect.get_height()
    
    ax.text(x + width/2, height + 2,
            str(height),
            fontsize=14,
            ha='center')
```
