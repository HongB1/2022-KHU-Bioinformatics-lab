````python
# chrom + chromStart + chromEnd -> Pos (chr:####-####)
# Histone.bed -> Histone
# 동일한 gene에서 같은 종류의 histone이 발견되면 distance가 적은 걸로, distance가 같으면 HsignalValue가 큰 값을 남기기

def reshape_df(df):
    df['Pos'] = 'tmp'
    df['start'] = df['start'].apply(lambda x: str(x))
    df['end'] = df['end'].apply(lambda x: str(x))
    
    Pos = df['chr'] + ':' + df['start'] + '-' + df['end']
    df['Pos'] = Pos

    df['Histone'] = df['Histone'].apply(lambda x : x.split('.')[0].split('_')[0])

    result = df.reindex(columns=['Pos', 'gene_name', 'distance with gene', 'Histone', 'HsignalValue'])
    return result
````

````python
# chrom + chromStart + chromEnd -> Pos (chr:####-####)
# Histone.bed -> Histone
# 동일한 gene에서 같은 종류의 histone이 발견되면 distance가 적은 걸로, distance가 같으면 HsignalValue가 큰 값을 남기기

def reshape_df(df):
    df['Pos'] = 'tmp'
    
    for idx in range(len(df)):
        df['Pos'][idx] = f'{df["chr"][idx]}:{df["start"][idx]}-{df["end"][idx]}'
        df['Histone'][idx] = df['Histone'][idx].split('.')[0].split('_')[0]

    result = df.reindex(columns=['Pos', 'gene_name', 'distance with gene', 'Histone', 'HsignalValue'])
    return result
````

````python
## 동일한 유전자에서 같은 histone이 발견되면 distance 적은 걸로, 같으면 HsignalValue가 큰 값을 남기기. 
## 같으면 둘중 하나 없애
## list에 넣어서 drop(list)

unique_gene_list = concatDF['gene_name'].unique()
df = pd.DataFrame()
for gene in unique_gene_list:
    GENE = concatDF[concatDF['gene_name']==gene]
    if len(GENE[GENE.duplicated(subset='Histone', keep=False)]) > 0 :
        unique_histone_list = GENE['Histone'].unique()
        for histone in unique_histone_list:
            HISTONE = GENE[GENE['Histone']==histone]
            if len(HISTONE[HISTONE['distance with gene'] == HISTONE['distance with gene'].min()]) > 0 :
                HISTONE_dis_min = HISTONE[HISTONE['distance with gene'] == HISTONE['distance with gene'].min()]
                HISTONE_dis_min_hsignal_max = HISTONE_dis_min[HISTONE_dis_min['HsignalValue'] == HISTONE_dis_min['HsignalValue'].max()]
                df = pd.concat([df, HISTONE_dis_min_hsignal_max]) 

            else : df  = pd.concat([df, HISTONE[HISTONE['distance with gene'] == HISTONE['distance with gene'].min()]])
    else : df = pd.concat([df, GENE])
````
