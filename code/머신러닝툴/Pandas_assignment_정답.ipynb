```python
# groupby를 이용해서 집계해보도록 하겠습니다.
# hobby를 기준으로 집계해봅시다.
## 먼저 DataFrameGroupBy 객체를 생성해보세요.
gp = df.groupby(by='hobby')
gp
```

```python
# 생성한 DataFrameGroupBy 객체의 agg() 메서드를 사용하여, sum, mean 순서로 집계해보도록 하겠습니다.
gp.agg(['sum', 'mean'])
```

```python
# 여기는 지우지 말고 실행시키세요.
df.reset_index(inplace=True, drop=False)
df
```

```python
# 이제 pandas.DataFrame.pivot_table 을 사용해보죠 
## index를 'hobby'로 정하고, 'BMI' 값에 대해 평균을 구해봅시다. (pivot_table 메소드의 aggfunc 파라미터의 기본값 인수는 평균을 구하는 것이죠)
df.pivot_table(index='hobby', values='BMI')
```

```python
# 이제 pandas.DataFrame.pivot_table 을 사용해보죠 
## index를 'hobby'로 정하고, 'BMI' 값의 합계를 구해봅시다. aggfunc 파라미터의 값을 바꿔주면 되겠죠?
df.pivot_table(index='hobby', values='BMI', aggfunc='sum')
```

```python
## index를 'hobby'로 정하고, 'BMI'와 'final_grade' 값의 평균을 구해봅시다.
df.pivot_table(index='hobby', values=['BMI', 'grade']) 
```

```python
## index를 'hobby'로 정하고, 'BMI'와 'final_grade' 값의 평균을 구할 것인데, over25 컬럼을 기준으로 수평 집계 해봅시다.
df.pivot_table(index='hobby', values=['BMI', 'grade'], columns='over 21')
```

```python
## ## index를 'hobby'로 정하고, 전체 value들의 평균을 구하고, margins 파라미터 설정을 통해 전체의 평균도 구해봅시다
df.pivot_table(df, index='hobby', margins=True)
```

```python
# 여러 값에 대해 여러 함수 적용 => list를 이용한 다중 함수 적용 (np.sum, np.mean)
# 엑셀의 피벗테이블과 유사하게 여러 값에 대해서 여러 값(values)에 대해 여러 함수(aggfunc)를 설정할 수 있습니다.
### values는 'BMI'와 'grade'에 대해 
### index는 'hobby'를 기준으로, 
### 컬럼은 'over 21'로 해서 
### aggfunc는 np.sum과 np.mean 두가지 함수를 적용해봅시다.
### margins도 설정해주어 전체에 대해서도 집계하겠습니다.
df.pivot_table(values=['BMI','grade'],index=['hobby'], columns='over 21',aggfunc=[np.sum,np.mean], margins=True)
```
